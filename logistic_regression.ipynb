```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
np.random.seed(0)
```

```python
hours_studied=np.random.randint(1,10,100)
pass_fail=np.where(hours_studied>5,1,0)
```

```python
plt.scatter(hours_studied,pass_fail,color='blue')
plt.xlabel('hours studied')
plt.ylabel('pass(1)/fail(0)')
plt.title('hours studied vs pass/fail')
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(hours_studied.reshape(-1,1),pass_fail,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(x_test,y_pred)
print(f"accuracy:{accuracy}")
```

```python
plt.scatter(x_test,y_test,color='green')
plt.plot(np.sort(x_test,axis=0),model.predict_proba(np.sort(x_test,axis=0))[:,])
plt.xlabel('hours studied')
plt.ylabel('pass probability')
plt.title('Logistic regression decision boundary')
plt.show()
```
